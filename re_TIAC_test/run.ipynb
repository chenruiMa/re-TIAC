```python
import torch
```

```python
a = torch.nn.Embedding(3,10)
```

```python
b = torch.nn.Embedding(3,10)
```

```python
soft = torch.nn.Softmax(dim=-1)
```

```python
a = a.weight
b = b.weight
```

```python
sim = torch.matmul(a,b.T)
sim = soft(sim)
```

```python
sim
```

```python
diag = torch.diag(sim)
diag = torch.diag_embed(diag)
diag
```

```python
after = torch.matmul(diag,b)
```

```python
torch.matmul(a,after.T)
```
